Cleaning the data

```python
import pandas as pd
import numpy as np
```

```python
ESK1 = pd.read_csv('EskwenilaEducationData.csv', delimiter = ';')
```

```python
ESK1.head()
```

```python
ESK1.columns
```

```python
#replace , to .
ESK1['Age <15, %'] = ESK1['Age <15, %'].str.replace(',','.')
ESK1['Age 16-25, %'] = ESK1['Age 16-25, %'].str.replace(',','.')
ESK1['Age 26-64, %'] = ESK1['Age 26-64, %'].str.replace(',','.')
ESK1['Age >65, %'] = ESK1['Age >65, %'].str.replace(',','.')
```

```python
#change type str to int/float
ESK1['Age <15, %'] = pd.to_numeric(ESK1['Age <15, %'])
ESK1['Age 16-25, %'] = pd.to_numeric(ESK1['Age 16-25, %'])
ESK1['Age 26-64, %'] = pd.to_numeric(ESK1['Age 26-64, %'])
ESK1['Age >65, %'] = pd.to_numeric(ESK1['Age >65, %'])
```

```python
ESK1.info()
```

```python
ESK1.info()
```

```python
ESK1 = ESK1.rename(columns={'Population, absolute':'Population','Age <15':'Age <15, %', 'Age 16-25':'Age 16-25, %', 'Age 26-64':'Age 26-64, %', 'Age >65':'Age >65, %'})
```

```python
ESK1.info()
```

```python
ESK1.to_csv('CLEAN_100.csv')
```
